```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
#with threshold 2.43
variabilityk3 = np.load("variabilityk3.npy")[2:] #removing first two so that variability starts from l=1 (data was calculated from l=0)
variabilityk4 = np.load("variabilityk4.npy")[2:]
activated_k3 = np.load("activated_k3.npy")[0:23]
activated_k4 = np.load("activated_k4.npy")[0:23]
```

```python
#calculating the correlation coefficient between proportion of activated nodes and variability as a function of L
nodes = 116
l_values = np.linspace(1,12,23)
prop_activated_k3 = activated_k3/nodes
prop_activated_k4 = activated_k4/nodes

corr_k3, p_value_k3 = pearsonr(variabilityk3, prop_activated_k3)
corr_k4, p_value_k4 = pearsonr(variabilityk4, prop_activated_k4)
print(f"Pearson correlation coefficient: {corr_k3:.2f}, P-value: {p_value_k3:.5f}")
print(f"Pearson correlation coefficient: {corr_k4:.2f}, P-value: {p_value_k4:.8f}")
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(l_values, variabilityk3, label="Variability K = 3")
ax.plot(l_values, prop_activated_k3, label="Proportion of activated nodes K = 3")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.title(f"Pearson correlation coefficient: {corr_k3:.2f}, P-value: {p_value_k3:.5f}", fontsize=14)
plt.xlabel("Internode coupling L", fontsize=12)
plt.ylabel("Order variablity and Proportion of activated nodes", fontsize=12)
plt.legend()
plt.savefig("corr_k3")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(l_values, variabilityk4, label="Variability K = 4")
ax.plot(l_values, prop_activated_k4, label="Proportion of activated nodes K = 4")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel("Internode coupling L", fontsize=12)
plt.ylabel("Order variablity and Proportion of activated nodes", fontsize=12)
plt.legend()
plt.title(f"Pearson correlation coefficient: {corr_k4:.2f}, P-value: {p_value_k4:.8f}", fontsize=14)
plt.savefig("corr_k4")
plt.show()
```
